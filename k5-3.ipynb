# Tipos: Objectos, Planetas, Atributos

Do ponto de vista computacional, a definição de objectos **variados** e a definição de **atributos** é feita, respetivamente, através dos **Tipos**
* $T = A + B + \ldots$ `// união disjunta`
* $T = A \times B \times \ldots$ `// registo`

Recorde-se que em **F#** definir um tipo $T = A + B + \ldots$ corresponde a escrever

```
type T = TagA of A | TagB of B | ...
```

e que definir um tipo $T = A \times B \times \ldots$ corresponde a escrever

```
type A = int
type B = float
type C = string
type T = { SelA: A; SelB: B; SelC: C }
```

```
{ SelA = 2; SelB = 2.0; SelC = "nome"}
```
